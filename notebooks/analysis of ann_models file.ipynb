```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../metrics/ann_models.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
for i in range(len(df)):
    df['Accuracy'][i] = str(df['Accuracy'][i]).replace("[","")
    df['Kappa'][i] = str(df['Kappa'][i]).replace("]","")
    
```

```python
df.info()
```

```python
df.head()
```

```python
df['Accuracy'] = df['Accuracy'].astype(float)
df['Kappa'] = df['Kappa'].astype(float)
```

```python
df.info()
```

```python
for i in range(len(df)):
    df['Epochs'][i] = str(df['Epochs'][i]).replace("for ","")
    df['Epochs'][i] = str(df['Epochs'][i]).replace(", the results are:","")
    df['Epochs'][i] = str(df['Epochs'][i]).replace(",","")
    
```

```python
df.head()
```

```python
df[['Epochs','Neurons']] = df.Epochs.str.split(expand=True)
```

```python
df.head()
```

```python
df['Epochs'] = df['Epochs'].astype(int)
df['Neurons'] = df['Neurons'].astype(int)
```

```python
df.info()
```

```python
df['Sums'] = df['Accuracy'] + df['F1 Score'] + df['Recall'] + df['Precision'] + df['MCC'] + df['Kappa']
```

```python
df.head()
```

```python
df = df.sort_values('Sums',ascending=False)
```
