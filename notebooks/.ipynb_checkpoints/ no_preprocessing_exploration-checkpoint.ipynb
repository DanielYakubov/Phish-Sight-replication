```python
import numpy as np
import warnings
from scipy import stats
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

# Model Imports
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# Metric imports
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, matthews_corrcoef,cohen_kappa_score, precision_recall_curve, PrecisionRecallDisplay, RocCurveDisplay
from sklearn.model_selection import GridSearchCV, train_test_split

# other
import pandas as pd
import joblib

```

```python
data=pd.read_csv("../data/scraped/all_data_scraped.csv")
data
```

```python
sns.countplot(x=data['status'])
```

```python
# So this is the data before preprocessing, we want to balance it to be more similar to pandey/mishra's balance
# But importantly, we do not want to preprocess and filter out links that had errors
```

```python
_, values_0 = data.status.value_counts()
values_0
```

```python
# getting classes
df_class_0 = data[data['status'] == 0]
df_class_1 = data[data['status'] == 1]
```

```python
# under sampling class 1, by the same proportion pandey/mishra had
```

```python
proportion = 3500/2700
adjusted_class_1 = df_class_1.sample(round(values_0*proportion))
```

```python
# now adjusting the data to the same proportion as pandey & mishra
adjusted_data = pd.concat([df_class_0, adjusted_class_1], axis=0)
sns.countplot(x=adjusted_data['status'])
```

```python
# one hot encoding
one_hot = pd.get_dummies(adjusted_data.brand_name)
adjusted_data = pd.concat([adjusted_data, one_hot], axis=1)
```

```python
adjusted_data.head()
```

```python
# we are not using any text based features since this is exactly pandey mishra. no_brand is redundant
adjusted_data = adjusted_data.drop(columns=['URL', 'text', 'no_brand', 'brand_name'], axis=1)
```

```python
adjusted_data.head()
```

```python
# splitting data
X = adjusted_data.drop(columns=['status'])
y = adjusted_data['status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# scaling data
# getting scaled features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix

def get_fpr(y_test, y_hat):
    tn, fp, fn, tp = confusion_matrix(y_test, y_hat).ravel()
    fpr = fp/(fp+tn)
    return fpr
```

```python
def print_eval(y_test, y_hat):
    acc = accuracy_score(y_test, y_hat)
    f1 = f1_score(y_test, y_hat)
    fpr = get_fpr(y_test, y_hat)
    recall = recall_score(y_test, y_hat)
    precision = precision_score(y_test, y_hat)
    mcc = matthews_corrcoef(y_test, y_hat)
    kappa = cohen_kappa_score(y_test, y_hat)
    print("accuracy", acc)
    print("F1", f1)
    print("fpr", fpr)
    print("recall", recall)
    print("precision", precision)
    print("MCC", mcc)
    print("Cohen's K", kappa)
    return acc, f1, fpr, recall, precision, mcc, kappa
```

```python
# training models, no tuning as we are replicating here
```

### Naive Bayes

```python
nb = GaussianNB()
nb.fit(X_train, y_train) # doesn't need standardized X_train
y_pred = nb.predict(X_test)
nb_metrics = print_eval(y_test, y_pred)
```

### Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
y_pred = lr.predict(X_test_scaled)
lr_metrics = print_eval(y_test, y_pred)
```

### SVC

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
y_pred = svc.predict(X_test_scaled)
svc_metrics = print_eval(y_test, y_pred)
```

### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train_scaled, y_train)
y_pred = dt.predict(X_test_scaled)
dt_metrics = print_eval(y_test, y_pred)
```

### Random Forest

```python
rf = RandomForestClassifier()
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_test_scaled)
rf_metrics = print_eval(y_test, y_pred)
```

```python
accs = []
fprs = []
recalls = []
models = ["Naive Bayes", "Logistic Regression", "SVM", "Decicion Tree", "Random Forest"]
for metrics in [nb_metrics, lr_metrics, svc_metrics, dt_metrics, rf_metrics]:
    acc, f1, fpr, recall, precision, mcc, kappa = metrics
    accs.append(acc)
    fprs.append(fpr)
    recalls.append(recall)
metrics_df = pd.DataFrame(data=zip(models, accs, fprs, recalls), columns=['model', 'accuracy', 'fpr', 'recall'])
metrics_df
```

```python
metrics_df.to_csv('../metrics/replication.csv', index=False)
```
