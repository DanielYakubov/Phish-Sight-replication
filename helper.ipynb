```python
import pandas as pd
```

```python
existing_links = pd.read_csv('data/scraped/all_data_scraped.csv')
```

```python
already_have = existing_links['URL'].to_list()
```

```python
already_have
```

```python
new = pd.read_csv('data/links/top-1m.csv', names=['idx', 'URL'])
```

```python
new.head()
```

```python
new = new["URL"].to_list()
```

```python
new
```

```python
# filtering
real_new = []
for n in new:
    if n not in already_have:
        real_new.append(n)
```

```python
len(real_new)
```

```python
labels = [0] * len(real_new)
```

```python
out_df = pd.DataFrame(data=zip(real_new, labels), columns=["URL", "status"])
```

```python
out_df.to_csv("data/links/additional_legit.csv", index=False)
```
